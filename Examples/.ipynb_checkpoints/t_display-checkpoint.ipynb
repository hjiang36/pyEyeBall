# Introduction to display class #
In this tutorial, we will go through attributes, properties and methods of display class.

## Import and module loading ##
The display class can be directly imported from pyEyeBall as

```python
%matplotlib inline
from pyEyeBall import Display
```

Alternatively, we can import using the actual full path as

```python
from pyEyeBall.Source.Objects.Display import Display
```

## Create display
In current version of pyEyeBall, display are usually created with ISETBIO .mat display calibration file. This is done by class method 'init_with_isetbio_mat_file'. As an example, we can create a Sony OLED by

```python
d = Display.init_with_isetbio_mat_file('OLED-Sony.mat')
```

A bunch of calibrated display files are stored in 'Data/Display' folder. To create these displays, only file name should work. Otherwise, full absolute path is required. The full list of files in 'Data/Display' can be retrieved by static function 'ls_display' of Display class

```python
Display.ls_display()
```

## Display description, plot and visualization ##
__str__ method is implemented for the Display to generate decription string for the display instance. Thus, display instance can be converted to string explicitly by str() or implicitly. For example, we can use following command on the display instance:

```python
print(d)
```

To generate plots for attributes and properties for display instance, we can call 'plot' method of Display class. In current version, there are four different types of plots supported: 'spd', 'gamma', 'invert gamma' and 'gamut'.

```python
d.plot('spd')
d.plot('gamut')
```

There is also a DisplayGUI class, especially for creating GUI for display instance. To show that GUI window, use command d.visualize(). The GUI window requires PyQt4 to run. PyQt4 does not support pip and thus need to be installed independently.

## Attributes and Properties ##
Fundamental attributes are those directly stored in display instance and can be directly read from and write to the object. The fundamental attributes for the display class are:
* name (str): name of display instance
* gamma (numpy.ndarray): gamma distortion table of the display of shape (n_levels, n_primaries)
* spd (numpy.ndarray): spectral power distribution of the primaries in shape (n_wave, n_primaries). This fields represents <strong>mean</strong> spectral power distribution at pixel level and it's not equivalent to the peak spd of the materials of primaries. spd are stored in energy units.
* dist (float): distance between display and observer in meters
* dpi (float): display spatial resolution in dots per inch. We assume that spatial resolution along horizontal and vertical directions are the same
* is_emissive (bool): is emissive display or reflective display, should always be True for current version
* ambient (numpy.ndarray): spectra power distribution of display at darkest level
* dixel (collections.namedtuple): pixel layout structure, contains .intensity_map, .control_map and n_pixels

```python
d.name = 'example display'
d.dist = 2.0
print(d.dpi)
```

As a special case, wavelength samples are also store directly in the display instance as <strong>_wave</strong>. But it should not be accessed directly. Instead, to read/write wavelength samples, use <strong>wave</strong> property instead. 

```python
import numpy as np
print(d.spd.shape)
d.wave = np.arange(400.0, 710.0, 10.0)
print(d.spd.shape)
```

It can be seen that setting the wavelength samples also invokes interpolation to the corresponding spectra data. Besides fundamental attributes, there are a lot more computed properties. These properties are readable, but not settable. Here are some of them and you can refer to the api document for the full list:

```python
print('Color depth: ' + str(d.n_bits) + ' bits')
print('Number of primaries: ' + str(d.n_primaries))
print('White point: ' + str(d.white_xyz))
print('Peak luminance: ' + str(d.peak_luminance))
```

Besides these properties, there are also some computed conversion matrix from the display. For example:

```python
print('XYZ for green primary' + str(np.dot([0, 1, 0], d.rgb2xyz)))
print('S cone isolating direction' + str(np.dot([0, 0, 1], np.linalg.inv(d.rgb2lms))))
```

<strong>Note: </strong> all the RGB values in the conversion matrix are referred as linear RGB. For digital values, we should use gamma table to find linear RGB value first. To convert full linear RGB image to xyz or equivalent srgb image, use d.compute_xyz() and d.compute_srgb()

## Gamma distortion of display
In display class, lookup_digital() method converts digital values to linear RGB values through gamma table. Similarly, lookup_linear() method to converts linear RGB values to digital values.

```python
print(d.lookup_digital(np.array([[[150, 130, 120]]])))
print(d.lookup_linear(np.array([[[0.5, 0.3, 0.2]]])))
```

In some situations, we would like to have a simulated display without gamma distortion. We could achieve that by

```python
d.plot('gamma')
d.gamma = np.tile(np.linspace(0, 1, d.n_levels), (3, 1)).T
d.plot('gamma')
```
