# Introduction to scene class #
In this tutorial, we will go through attributes, properties and methods of scene class.

## Import and module loading ##
The scene class can be directly imported from pyEyeBall as

```python
%matplotlib inline
from pyEyeBall import Scene
```

Alternatively, we can import using the actual full path as

```python
from pyEyeBall.Source.Objects.Scene import Scene
```

## Create Scene ##
There are two ways to create a scene:
* By name of the preset scene
* By specifying image on a calibrated display

In current version of pyEyeBall, there are two preset scenes that can be created by name: 'macbeth' and 'noise'. For macbeth, it creates a macbeth color checker with desired patch size and specified illuminate. For noise pattern, the program creates random gaussian white noise with specified size.

```python
scene = Scene('macbeth')
scene.plot('srgb')
scene = Scene('noise')
```

As an alternative to scene creation by name, we can define the scene by specifying an RGB image on a calibrated display. In this case, RGB image should be of values between 0~1.

```python
from PIL import Image
import requests
import numpy as np
from pyEyeBall import Display
from io import BytesIO

response = requests.get("http://scarlet.stanford.edu/validation/SCIEN/RGB/LStryer/rocksWater.jpg")
image = np.array(Image.open(BytesIO(response.content))).astype(float)/255.0

scene = Scene.init_with_display_image(Display.init_with_isetbio_mat_file("OLED-Sony.mat"), image)  # create scene
scene.plot('srgb')
```

## Display description, plot and visualization ##
str method is implemented for the Scene class to generate decription string for the scene radiance. Thus, scene instance can be converted to string explicitly by str() or implicitly. For example, we can use following command on the scene instance:

```python
scene = Scene()
print(scene)
```

To generate plots for attributes and properties for scene instance, we can call 'plot' method of Scene class. In current version, there are three different types of plots supported: 'illuminant energy', 'illuminant photons', 'srgb'.

```python
scene.plot('illuminant energy')
```

There is also a SceneGUI class, especially for creating GUI for scene instance. To show that GUI window, use command scene.visualize(). The GUI window requires PyQt4 to run. PyQt4 does not support pip and thus need to be installed independently.

## Attributes and Properties ##
Fundamental attributes are those directly stored in scene instance and can be directly read from and write to the object. The fundamental attributes for the scene class are:
* name (str): name of the Scene instance
* photons (numpy.ndarray): scene spectra radiance data
* fov (float): horizontal field of view of the scene in degrees
* dist (float): the distance between the scene and the observer
* illuminant (pyEyeBall.Illuminant): illumination properties

```python
print(scene.name)
scene.fov = 2.0
```

<strong>Note: </strong> wavelength samples is not a fundamental attribute in scene instance. Instead, scene wavelength is defined by the wavelength in its illuminant structure. For external usage, we build a property 'wave' for scene instance, which can be read/write just as normal attribute.

```python
print(scene.photons.shape)
scene.wave = np.arange(400.0, 710.0, 5.0)
print(scene.photons.shape)
```

It can be seen that setting the wavelength samples also invokes interpolation to the corresponding spectra data. 

Besides fundamental attributes, there are a lot more computed properties. These properties are readable, but not settable. Here are some of them and you can refer to the api document for the full list:

```python
print('Shape: ' + str(scene.shape))
print('width: ' + str(scene.width) + 'meters')
print('height: ' + str(scene.height) + 'meters')
print('mean_luminance: ' + str(scene.mean_luminance) + 'cd/m2')
```

Among these properties, mean luminance is a special case because it's adjustable. When setting mean luminance to a new value, we scale the scene radiance data to match the desired mean luminance.

```python
scene.mean_luminance = 200.0
```

Besides these properties, there are also methods in scene class that converts the radiance map into values in different color space. For example:

```python
print(scene.xyz.shape)
import matplotlib.pyplot as plt
plt.imshow(scene.srgb)
plt.show()
```
